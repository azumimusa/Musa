```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs=np.array([1,2,3,4,5,6], datatype=np.float64)
ys=np.array([5,4,6, 5,6,7], datatype=np.float64)
```

```python
plt.scatter(xs,ys)
plt.show()
```

```python
def best_fit_slope_intercept(xs, ys):
    m=((mean(xs)*mean(ys))-(mean(xs*ys)))/(((mean(xs))**2) - (mean(xs**2)))
    b=mean(ys) - m*mean(xs)
    return m,b
m, b =best_fit_slope_intercept(xs, ys)
def square_error(ys_origin, ys_line):
    return sum((ys_line -ys_origin)**2)
def coefficient_of_determination(ys_origin, ys_line):
    y_mean_line=[mean(ys_origin) for y in ys_origin]
    square_error_reg=square_error(ys_origin, ys_line)
    square_error_y_mean=square_error(ys_origin, y_mean_line)
    return (1 - square_error_reg/square_error_y_mean)
```

```python
print(m , b)
```

```python
regression_line=[(m*x)+ b for x in xs]
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
```

```python
predict_x = 10
predict_y=(m*predict_x) - b
```
